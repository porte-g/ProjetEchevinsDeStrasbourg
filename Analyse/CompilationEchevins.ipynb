Ce script permet de compiler les occurences des échevins des métiers des listes du conseil en un seul document.

```python
import pandas as pd
import os
```

```python
dfComp = pd.DataFrame(columns = ['Unnamed: 0','id', 'anneeMagistrature', 'magistrature', 'occurence', 'prenom', 'nom', 'surnom', 'genName', 'titres_roles', 'corps civique', 'corporation', 'poele', 'page', 'colonne', 'ligne'])
for filename in os.listdir('./../Sources/ListesConseil'):
    if filename.endswith('.csv'):
        with open(os.path.join('./../Sources/ListesConseil', filename)) as f:
            df = pd.read_csv(f)
            entry = df.loc[df['corps civique'] == 'echevinat']
            dfComp = dfComp.append([entry])
    dfComp.to_csv('./Data/OccEch.csv',index=False)
```
