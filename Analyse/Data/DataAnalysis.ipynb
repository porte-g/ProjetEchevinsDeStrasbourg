```python
df = pd.read_csv('./../../Sources/ListeArmement/1444ListeArmement.csv')
```

```python
df1 = df.groupby('corporation').count()
df1 = df1.sort_values(by=['id'], ascending=False)
df1
```

```python
plot = df1['id'].plot.pie(figsize=(12,12))
plot.figure.savefig('./../../imagesMD/echevinscorpo.png')
```

```python
labels = df1.index.to_list()
sizes = df1['id'.to_list()

fig1, ax1 = plt.subplots()

ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
