```python
def ReconciliationDF(InputDf, OutputDf):
    """
    Cette fonction crée un tableur de réconciliation brut à partir de la matrice de similarité d'un tableur.
    """
    df = pd.read_csv(InputDf, index_col=0)
    listemax =[]
    for i in df.index :
        sousliste =[i]
        #print(i)
        for j in df.head():
            if i != j :
                if df.at[i, j] > 0.95 :
                    sousliste.append(j)
        listemax.append(sorted(sousliste))

    len(listemax)

    listemax2 = []
    for i in listemax:
        if i not in listemax2:
            listemax2.append(i)

    listemax3 = []
    for element in listemax2 :
        dico ={}
        dico['ids'] = element
        dico['length'] = len(element) 
        listemax3.append(dico)

    df2= pd.DataFrame(list(listemax3))
    df2.to_csv(OutputDf)
```

```python
def Rec2Auth(dfRec, dfOcc, Output):
    """
    Cette fonction crée un tableur d'autorité (Dfauth.csv), à partir d'un tableur de réconciliation et d'un tableur d'occurences.
    """
    df2 = pd.read_csv(dfRec, converters={"ids": literal_eval})
    for index, row in df2.iterrows():
        df2.at[index,'length'] = len(df2.at[index,'ids'])
    df2['length'] = df2['length'].astype(int)

    RefEch = pd.read_csv(dfOcc, index_col = None)
    RefEch = RefEch.set_index('id')
    RefEch.loc[:,'anneeMagistrature'] = RefEch.loc[:,'anneeMagistrature'].apply(int)

    ListeDicosPersonnes = []

    n=1
    for element in df2['ids']:

        DicoPersonne = {}

        length = len(str(n))
        nID = str(n).zfill(4)
        DicoPersonne['Recid'] =  'aves_aa_4R_auth_' + nID
        n = n + 1
        
        DicoPersonne['subid'] = []
        DicoPersonne['Unnamed: 0'] = []
        DicoPersonne['anneeMagistrature'] = []
        DicoPersonne['magistrature'] = []
        DicoPersonne['occurence'] = []
        DicoPersonne['prenom'] = []
        DicoPersonne['nom'] = []
        DicoPersonne['surnom'] = []
        DicoPersonne['genName'] = []
        DicoPersonne['titres_roles'] = []
        DicoPersonne['corps_civique'] = []
        DicoPersonne['corporation'] = []
        DicoPersonne['poele'] = []
        DicoPersonne['page'] = []
        DicoPersonne['colonne'] = []
        DicoPersonne['ligne'] = []


        for subelement in element:

            DicoPersonne['subid'].append(subelement)
            DicoPersonne['Unnamed: 0'].append(RefEch.at[subelement, 'Unnamed: 0'])
            DicoPersonne['anneeMagistrature'].append(RefEch.at[subelement, 'anneeMagistrature'])
            DicoPersonne['magistrature'].append(RefEch.at[subelement, 'magistrature'])
            DicoPersonne['occurence'].append(RefEch.at[subelement, 'occurence'])
            DicoPersonne['prenom'].append(RefEch.at[subelement, 'prenom'])
            DicoPersonne['prenom'] = pd.Series(DicoPersonne['prenom']).fillna('').tolist()
            DicoPersonne['nom'].append(RefEch.at[subelement, 'nom'])
            DicoPersonne['nom'] = pd.Series(DicoPersonne['nom']).fillna('').tolist()
            DicoPersonne['surnom'].append(RefEch.at[subelement, 'surnom'])
            DicoPersonne['genName'].append(RefEch.at[subelement, 'genName'])
            DicoPersonne['titres_roles'].append(RefEch.at[subelement, 'titres_roles'])
            DicoPersonne['corps_civique'].append(RefEch.at[subelement, 'corps civique'])
            DicoPersonne['corporation'].append(RefEch.at[subelement, 'corporation'])
            DicoPersonne['poele'].append(RefEch.at[subelement, 'poele'])
            DicoPersonne['page'].append(RefEch.at[subelement, 'page'])
            DicoPersonne['colonne'].append(RefEch.at[subelement, 'colonne'])
            DicoPersonne['ligne'].append(RefEch.at[subelement, 'ligne'])
        
        sep = '?'    
        DicoPersonne['AuthNameForename'] = "?".join(DicoPersonne['nom']).split(sep, 1)[0] + ", " + "?".join(DicoPersonne['prenom']).split(sep, 1)[0]

        ListeDicosPersonnes.append(DicoPersonne)
        Dfauth = pd.DataFrame(ListeDicosPersonnes)
        Dfauth['Length'] = Dfauth['subid'].str.len()
    Dfauth.to_csv(Output, index=False)
    return(Dfauth)
```
