```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./Data/LargeData/jaro_winkler_similarity.csv', index_col=0)
listemax =[]
for i in df.index :
    sousliste =[i]
    #print(i)
    for j in df.head():
        if i != j :
            if df.at[i, j] > 0.95 :
                sousliste.append(j)
    listemax.append(sorted(sousliste))
    
len(listemax)

listemax2 = []
for i in listemax:
    if i not in listemax2:
        listemax2.append(i)

listemax3 = []
for element in listemax2 :
    dico ={}
    dico['ids'] = element
    dico['length'] = len(element) 
    listemax3.append(dico)

df2= pd.DataFrame(list(listemax3))
df2.to_csv('./Data/ConsRec.csv')
```
